### ✏️ 1.1 Dessiner votre space invader

Il faut repérer les coordonnées des pixels. Pour cela, dessine à la main, sur une feuille quadrillée ton space invader, afin de bien repérer les coordonnées des pixels. 

Par exemple, la première ligne du space invader vert contient les pixels (2,7) et (8,7) :



Exécute le code ci-dessous, qui va afficher les trois premières lignes.

````python
couleur = "vert"

# première ligne :
dessine_pixel(2, 7, couleur)
dessine_pixel(8, 7, couleur)

# deuxième ligne :
dessine_pixel(0, 6, couleur)
dessine_pixel(3, 6, couleur)
dessine_pixel(7, 6, couleur)
dessine_pixel(10, 6, couleur)

# troisième ligne :
dessine_pixel(0, 5, couleur)
dessine_pixel(2, 5, couleur)
dessine_pixel(3, 5, couleur)
dessine_pixel(4, 5, couleur)
dessine_pixel(5, 5, couleur)
dessine_pixel(6, 5, couleur)
dessine_pixel(7, 5, couleur)
dessine_pixel(8, 5, couleur)
dessine_pixel(10, 5, couleur)

afficher_dessin()
````

On constate, pour la troisième ligne, que le code est très répétitif :
```python
dessine_pixel(0, 5, couleur)
dessine_pixel(2, 5, couleur)
dessine_pixel(3, 5, couleur)
dessine_pixel(4, 5, couleur)
dessine_pixel(5, 5, couleur)
dessine_pixel(6, 5, couleur)
dessine_pixel(7, 5, couleur)
dessine_pixel(8, 5, couleur)
dessine_pixel(10, 5, couleur)
```

On constate que seul le chiffre des abscisses est modifié à chaque ligne. Notons le **k**, chaque ligne est en fait :

```python
dessine_pixel(k, 5, couleur)
```
avec k qui vaut succesivement 0, puis 2, puis 3 ...

On peut remplacer les 9 lignes par seulement deux lignes en faisant une boucle :
```python
for k in [0, 2, 3, 4, 5, 6, 7, 8] :
    dessine_pixel(k, 5, couleur)
```



<div style="margin:2px;border:1px solid black;padding:10px;">

+ Une structure itérative ou boucle permet de répéter un ensemble d'instructions.
```python
    # cette boucle répète deux fois les instructions
    for _ in range(2):
        instructions
```
+ On utilise une collection de valeurs pour indiquer le nombre d'itérations (ou passage dans la boucle) :
```python
    # cette boucle répète trois fois les instructions
    for i in [1, 2, 3]:
        instructions
```
+ Dans ce cas, la variable ``i`` changera de valeur à chaque itération et prendra tour à tour les valeurs de la collection.
</div>

On peut donc créer une structure itérative pour chaque ligne, et notre code devient :

````python
# dessin complet du space invader
couleur = "vert"

# première ligne :
for k in [2, 8] :
    dessine_pixel(k,7, couleur)
# deuxième ligne :
for k in [0, 3, 7, 10] :
    dessine_pixel(k,6, couleur)
# troisième ligne :
for k in [0, 2, 3, 4, 5, 6, 7, 8, 10] :
    dessine_pixel(k,5, couleur)
# 4ème ligne :
for k in [0, 1, 2, 4, 5, 6, 8, 9, 10] :
    dessine_pixel(k,4, couleur)
# 5ème ligne :
for k in [1, 2, 3, 4, 5, 6, 7, 8, 9] :
    dessine_pixel(k,3, couleur)
# 6ème ligne :
for k in [2, 3, 4, 5, 6, 7, 8] :
    dessine_pixel(k,2, couleur)
# 7ème ligne :
for k in [3, 8] :
    dessine_pixel(k, 1, couleur)
# 8ème ligne :
for k in [2, 3, 8, 9] :
    dessine_pixel(k, 0, couleur)
        
afficher_dessin()
````

# 4. Dessiner un second space invader

On veux dessinner un second space invader, mais décalé de 15 pixels vers la droite.

<p>Si on considère la seconde ligne de mon space invader&nbsp;:</p>

```python
for k in [2, 3, 4, 5, 6] :
    dessine_pixel(k, 2)
```

Il y a deux méthodes principales pour dessiner cette même ligne 15 pixels sur la droite&nbsp;:

+ ajouter 15 à toutes les valeurs de la liste des valeurs de k

```python
for k in [17, 18, 19, 20, 21] :
    dessine_pixel(k, 2)
    
```
+ ajouter directement 15 à k dans l&#8217;instruction qui permet de dessiner un pixel

```python 
for k in [2, 3, 4, 5, 6] :
    dessine_pixel(k + 15, 2)
```

La seconde est la plus rapide, c&#8217;est celle que nous utiliserons par la suite.
